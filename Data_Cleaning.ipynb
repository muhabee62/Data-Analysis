```python
############# WRITE YOUR CODE IN THIS CELL (IF APPLICABLE)####################
import pandas as pd
# Importing and reading dataset from directory
data = pd.read_csv("C:/Users/muhab/Downloads/adult.csv")
df = pd.DataFrame(data)

df.info() # Checking the data information (important for getting a brief overview)
print("\n")

# Summing up all the null values in the columns
null_val_count = df.isna().sum()

# Getting a list of columns with null values
null_columns = null_val_count[null_val_count > 0].index.tolist()

print("The column with null values: ")
print(null_columns)
print("\nThe number of null values in each column: ")
print(null_val_count)

# Setting the display settings
pd.set_option('display.max_rows', 100)

print("\nThe first 100 rows of the Dataset: ")
print(df.head(100))

# Finding columns with data considered to be wrong
wrong_columns = df.columns[df.isin(['?']).any()]

print('\n The columns with wrong data: ')
print(wrong_columns)
```

```python
############# WRITE YOUR CODE IN THIS CELL (IF APPLICABLE)####################
# Identifying columns containing null values
columns_with_null = df.columns[df.isna().any()].tolist()

# Determining the data type of each column
data_types = {}
for column in columns_with_null:
    data_type = df[column].dtype
    if data_type == 'object':
        data_types[column] = 'Nominal data type'
    elif data_type == 'int64':
        data_types[column] = 'Numerical data type'
    else:
        data_types[column] = 'Ordinal data type'
        
print('The data type of columns with null values: ')
#b = []
for column, dtype in data_types.items():
    print(f'{column}: {dtype}')
    #b.append(b)
#print(b)

# Deleting the rows containing null values:
df_clean = df.dropna()

# Getting the number of null values in all columns
null_clean = df_clean.isna().sum()

print('\nThe number of null values in each column after deleting rows with null values: ')
print(null_clean)

# Importing and reading dataset from directory
data = pd.read_csv("C:/Users/muhab/Downloads/adult.csv")
df = pd.DataFrame(data)

# Finding the categorical columns with null values
cat_col_null =[]
for col in df.columns:
    if df[col].dtype == 'object' and df[col].isna().any():
        cat_col_null.append(col)

print('\nThe categorical columns with null values are: ')
print(cat_col_null)

# Filling the null values with the mode for each categorical column
mode_values = []
for column in cat_col_null:
    mode_value = df[column].mode()[0]
    df[column].fillna(mode_value, inplace=True)
    mode_values.append(mode_value)

print('\nThe mode of the categorical variables containing null values respectively are: ')
print(mode_values)    
print('\nThe dataframe after filling the null values with the mode of the columns:')
print(df)
```

```python
############# WRITE YOUR CODE IN THIS CELL (IF APPLICABLE)####################
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv("C:/Users/muhab/Downloads/adult.csv")
df = pd.DataFrame(data)

# Initializing the Label Encoder
label_encoder_instance = LabelEncoder()

# Encode the income column
df['income_encoded'] = label_encoder_instance.fit_transform(df['income'])

# Setting the display settings
pd.set_option('display.max_rows', 100)

print('The income column before and after encoding is displayed below : ')
print(pd.concat([df['income'], df['income_encoded']], axis=1))

unique_values = sorted(df['hours-per-week'].unique())
print('\n\n\nThe unique ordinal values in the "hours-per-week" variable are: ')
print(unique_values)

# Create a dictionary mapping each unique value to its rank
value_rank_mapping = {value: rank for rank, value in enumerate(unique_values, start=1)}

# Encode the hours-per-week column based on the rank mapping
df['hours-per-week-encoded'] = df['hours-per-week'].map(value_rank_mapping)

print("\n\n\nThe hours-per-week column before and after encoding is displayed below: ")
print(pd.concat([df['hours-per-week'], df['hours-per-week-encoded']], axis=1))

# Initializing the LabelEncoder
label_encoder = LabelEncoder()

# Encoding the 'education' column using LabelEncoder
df['education_encoded'] = label_encoder.fit_transform(df['education'])

print('\n\n\nThe education column before and after encoding is displayed below: ')
print(pd.concat([df[['education']], df['education_encoded']], axis=1))

data = pd.read_csv("C:/Users/muhab/Downloads/adult.csv")
df = pd.DataFrame(data)

# Appending the encoded 'education' column to the original Dataset
df['education_encoded'] = label_encoder.fit_transform(df['education'])

print("\n\n\nThe dataset after appending the encoded 'education' column:")
print(df)

from sklearn.preprocessing import StandardScaler

data = pd.read_csv("C:/Users/muhab/Downloads/adult.csv")
df = pd.DataFrame(data)

# Selecting the numerical columns
numerical_columns = df.select_dtypes(include=['int']).columns

# Initializing StandardScaler
scaler = StandardScaler()

# Normalizing the numerical columns using standard scaling
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

print("\n\n\nThe dataset after normalization of numerical columns:")
print(df)

```
