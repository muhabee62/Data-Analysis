## TASK : Apply four classifiers, i.e. Support Vector Machine (SVM), Decision Tree (DT), Random Forest (RF), and K-nearest neighbours (KNN) classifiers to the adult_WS#3 dataset  to predict the income column. Calculate the confusion matrix and evaluation metrics for all classifiers. Plot the features' importance values, determine the three most important features (i.e. columns) which have the highest impact on the income and report them in the report cell. 


```python
############# WRITE YOUR CODE IN THIS CELL (IF APPLICABLE)####################
# Importing the necessary libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import matplotlib.pyplot as plt

# Loading the dataset
df = pd.read_csv('C:/Users/muhab/Downloads/adult_WS#3.csv')

# Preprocessing the dataset
df.dropna(inplace=True)  # Deleting all rows containing null values
df.reset_index(drop=True, inplace=True)  # Re-indexing the resulting dataset

# Separating the input (features) into numerical and categorical columns
numerical_cols = df.select_dtypes(include=['int64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
categorical_cols = categorical_cols.drop('income')

# Setting the pipelines for the normalization of numerical and categorical data
num_pipe = Pipeline([('scaler', StandardScaler())])
cat_pipe = Pipeline([('encoder', OneHotEncoder())])

# Combining the processed numerical and categorical columns
combine_pipe = ColumnTransformer(transformers=[('num', num_pipe, numerical_cols), 
                                                ('cat', cat_pipe, categorical_cols)])

# Defining the final pipeline
pipeline = Pipeline(steps=[('combine_pipe', combine_pipe)])

# Fitting and transforming the data
df_processed = pipeline.fit_transform(df)

# Separating features and target variable
X = df_processed
y = df['income']

# Encoding the target variable
le = LabelEncoder()
y = le.fit_transform(y)

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initializing classifiers
svm_clf = SVC()
dt_clf = DecisionTreeClassifier()
rf_clf = RandomForestClassifier()
knn_clf = KNeighborsClassifier()

# Training the classifiers
svm_clf.fit(X_train, y_train)
dt_clf.fit(X_train, y_train)
rf_clf.fit(X_train, y_train)
knn_clf.fit(X_train, y_train)

# Making predictions
svm_preds = svm_clf.predict(X_test)
dt_preds = dt_clf.predict(X_test)
rf_preds = rf_clf.predict(X_test)
knn_preds = knn_clf.predict(X_test)

# Creating the confusion matrices for all classifiers' predictions
cm1 = confusion_matrix(y_test, svm_preds, labels=svm_clf.classes_)
cm2 = confusion_matrix(y_test, rf_preds, labels=rf_clf.classes_)
cm3 = confusion_matrix(y_test, knn_preds, labels=knn_clf.classes_)
cm4 = confusion_matrix(y_test, dt_preds, labels=dt_clf.classes_)

# Plotting confusion matrices
class_names = ['<=50K', '>50K']

plt.figure(figsize=(20, 10))

plt.subplot(2, 2, 1)
disp = ConfusionMatrixDisplay(confusion_matrix=cm1, display_labels=svm_clf.classes_)
disp.plot(cmap=plt.cm.Blues, ax=plt.gca())
plt.title("SVM")

plt.subplot(2, 2, 2)
disp = ConfusionMatrixDisplay(confusion_matrix=cm2, display_labels=rf_clf.classes_)
disp.plot(cmap=plt.cm.Blues, ax=plt.gca())
plt.title("RF")

plt.subplot(2, 2, 3)
disp = ConfusionMatrixDisplay(confusion_matrix=cm3, display_labels=knn_clf.classes_)
disp.plot(cmap=plt.cm.Blues, ax=plt.gca())
plt.title("KNN")

plt.subplot(2, 2, 4)
disp = ConfusionMatrixDisplay(confusion_matrix=cm4, display_labels=dt_clf.classes_)
disp.plot(cmap=plt.cm.Blues, ax=plt.gca())
plt.title("DT")

plt.tight_layout()
plt.show()

# This function takes the confusion matrix (cm) from the cell above and produces all evaluation matrix
def confusion_metrics(conf_matrix):
    TP = conf_matrix[1][1]
    TN = conf_matrix[0][0]
    FP = conf_matrix[0][1]
    FN = conf_matrix[1][0]
    print('True Positives:', TP)
    print('True Negatives:', TN)
    print('False Positives:', FP)
    print('False Negatives:', FN)
    
    # calculate accuracy
    conf_accuracy = (float(TP + TN) / float(TP + TN + FP + FN))
    
    # calculate mis-classification
    conf_misclassification = 1 - conf_accuracy
    
    # calculate the sensitivity
    conf_sensitivity = (TP / float(TP + FN))
    # calculate the specificity
    conf_specificity = (TN / float(TN + FP))
    
    # calculate precision
    conf_precision = (TN / float(TN + FP))
    # calculate f_1 score
    conf_f1 = 2 * ((conf_precision * conf_sensitivity) / (conf_precision + conf_sensitivity))
    print('-' * 50)
    print(f'Accuracy: {round(conf_accuracy, 2)}') 
    print(f'Mis-Classification: {round(conf_misclassification, 2)}') 
    print(f'Sensitivity: {round(conf_sensitivity, 2)}') 
    print(f'Specificity: {round(conf_specificity, 2)}') 
    print(f'Precision: {round(conf_precision, 2)}')
    print(f'f_1 Score: {round(conf_f1, 2)}')

# Printing the evaluation metrics for all classifiers
print('SVM metrics\n')
confusion_metrics(cm1)
print('\n\n')
print('RF metrics\n')
confusion_metrics(cm2)
print('\n\n')
print('KNN metrics\n')
confusion_metrics(cm3)
print('\n\n')
print('DT metrics\n')
confusion_metrics(cm4)
print('\n\n')

print(categorical_cols)

# Get feature importance values from the trained RandomForestClassifier
feature_importances = rf_clf.feature_importances_

# Get column names after encoding
encoded_categorical_features = pipeline.named_steps['combine_pipe'].transformers_[1][1].named_steps['encoder'].get_feature_names_out(input_features=list(categorical_cols))

encoded_feature_names = np.concatenate((numerical_cols, encoded_categorical_features))

# Create a DataFrame to hold feature importances
feature_importance_df = pd.DataFrame({'Feature': encoded_feature_names, 'Importance': feature_importances})

# Sort the DataFrame by importance values
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Plot the three most important features
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'][:3], feature_importance_df['Importance'][:3], color='skyblue')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Three Most Important Features')
plt.gca().invert_yaxis()  # Invert y-axis to have the most important features at


```
